```python
import os
from pathlib import Path
dataset_path = ['UAV--3/train/labels','UAV--3/test/labels']
for path in dataset_path:

    input_folder = Path(path)
    output_folder = input_folder.parent / (input_folder.name + '-new')

    # create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        if filename.endswith(".txt"):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)

            with open(input_path, "r") as f:
                lines = f.readlines()

            new_lines = []
            for line in lines:
                line = line.strip()
                if not line:
                    continue

                parts = line.split()
                parts[0] = "0"   # change class id to 0
                new_lines.append(" ".join(parts))

            with open(output_path, "w") as f:
                f.write("\n".join(new_lines))

print("All label files converted successfully.")
```

```python
import os
import shutil

# dataset_path = ['UAV--3/train/labels', 'UAV--3/test/labels']

for path in dataset_path:
    new_path = path + '-new'
    # Delete original folder if it exists
    if os.path.exists(path):
        shutil.rmtree(path)
        print(f"Deleted folder: {path}")
    # Rename the '-new' folder if it exists
    if os.path.exists(new_path):
        shutil.move(new_path, path)
        print(f"Renamed {new_path} --> {path}")
    else:
        print(f"{new_path} not found, skipping.")

print("Operation complete.")
```

```python
import yaml

yaml_path = 'UAV--3/data.yaml'  # Adjust to your file's location

# Read in the YAML file
with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Change nc and names
data['nc'] = 1
data['names'] = [0]

# Write back to the YAML file
with open(yaml_path, 'w') as f:
    yaml.safe_dump(data, f, sort_keys=False)

print("Updated data.yaml with nc: 1 and names: [0]")
```
