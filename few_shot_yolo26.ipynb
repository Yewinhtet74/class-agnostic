```python
from ultralytics import YOLO

model = YOLO("runs/detect/yolo26-visdrone-agnostic/weights/best.pt")
```

```python
model.train(
    data="DIOR/dior_filtered_classes/data.yaml", 
    imgsz=1280, 
    epochs=150, 
    dropout=0.2,
    name="yolo26-few_shot",
    resume=True)
```

```python
results = model.val(    data="DIOR/dior_filtered_classes/data.yaml", split='test')
```

```python
from ultralytics import YOLO
import os

# Paths
img_path = 'DIOR/dior_filtered_classes/test/images'
label_path = 'DIOR/dior_filtered_classes/test/labels'
model_path = 'runs/detect/yolo26-few_shot/weights/best.pt'
save_dir = 'runs/detect/predicted'

# Create save directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# Load YOLOv8 model
model = YOLO(model_path)

# Detect on all images and save results
results = model(source=img_path, save=True, project='runs/detect', name='predicted')

print(f'Detection complete. Results saved to {save_dir}')
```
