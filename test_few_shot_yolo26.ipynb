```python
from ultralytics import YOLO
import os
import cv2
import glob

# Paths
img_path = 'UAV--3/test/images'
label_path = 'UAV--3/test/labels'
model_path = 'runs/detect/yolo26-few_shot/weights/best.pt'
save_dir = 'runs/detect/predicted_with_gt'
os.makedirs(save_dir, exist_ok=True)

# Load YOLOv8 model
model = YOLO(model_path)

# Get all image files
image_files = glob.glob(os.path.join(img_path, '*'))

for img_file in image_files:
    img_name = os.path.basename(img_file)
    label_file = os.path.join(label_path, os.path.splitext(img_name)[0] + '.txt')

    # Run detection
    results = model(img_file)
    pred_img = results[0].plot()  # contains prediction boxes

    # Draw GT if label file exists
    img_cv = cv2.imread(img_file)
    h, w = img_cv.shape[:2]

    # Parse GT bounding boxes
    if os.path.exists(label_file):
        with open(label_file, 'r') as lf:
            for line in lf:
                parts = line.strip().split()
                if len(parts) >= 5:
                    cls, x, y, bw, bh = map(float, parts[:5])
                    # Convert normalized xywh to pixel coords
                    cx, cy = x * w, y * h
                    bw_pix, bh_pix = bw * w, bh * h
                    x1, y1 = int(cx - bw_pix / 2), int(cy - bh_pix / 2)
                    x2, y2 = int(cx + bw_pix / 2), int(cy + bh_pix / 2)
                    # Draw green rectangle for GT: BGR (0,255,0), thickness 2
                    cv2.rectangle(pred_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Save result with predictions + GT
    out_path = os.path.join(save_dir, img_name)
    cv2.imwrite(out_path, pred_img)

    print(f"Processed {img_name}")

print(f"All images saved to {save_dir}")
```
