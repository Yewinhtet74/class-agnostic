```python
from ultralytics import YOLO
import os

# Paths
img_path = 'UAV--1/test/images'
model_path = 'runs/detect/yolo26-visdrone-agnostic/weights/best.pt'
save_dir = 'runs/detect/predicted'

# Create save directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# Load YOLOv8 model
model = YOLO(model_path)

# Detect on all images and save results
results = model(source=img_path, save=True, project='runs/detect', name='predicted')

print(f'Detection complete. Results saved to {save_dir}')
```
