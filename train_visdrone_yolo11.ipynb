```python
from ultralytics import YOLO

model = YOLO("yolo11n.yaml")
model.train(
    data="VisDrone/data.yaml", 
    imgsz=1280, 
    epochs=150, 
    patience=15, 
    project="yolo11n-visdrone-agnostic",
    name="training",
    optimizer='Adam')
```

```python
results = model.val(data='VisDrone/data.yaml', split='test')
```
