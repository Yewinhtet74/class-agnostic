```python
from ultralytics import YOLO

model = YOLO("yolo26n.yaml")
model.train(
    data="VisDrone/data.yaml", 
    imgsz=1280, 
    epochs=100, 
    patience=15, 
    dropout=0.1,
    project="yolo26-visdrone-agnostic",
    name="training",
    optimizer='Adam')
```

```python
results = model.val(data='VisDrone/data.yaml', split='test')
```
